```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import StandardScaler

ttswing = pd.read_csv("./assignTTSWING.csv")
```

```python
# 清除异常值（'???'）所在的行
for col in ['age', 'height', 'weight', 'playYears']:
    if col in ttswing.columns:
        ttswing = ttswing[ttswing[col] != '???']
```

```python
# === 提取特征 X 和标签 y ===
X = ttswing.drop(columns=['testmode','teststage'])
y = ttswing['testmode']

X = X.astype({col: int for col in X.select_dtypes('bool').columns})

# === 特征标准化===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === 计算互信息 ===
mi_scores = mutual_info_classif(X_scaled, y, discrete_features='auto')
mi_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

# === 互信息得分最高的特征 ===
plt.figure(figsize=(10, 6))
mi_series.plot(kind='barh', color='steelblue')
plt.gca().invert_yaxis()
plt.title("Features by Mutual Information with 'testmode'")
plt.xlabel("Mutual Information Score")
plt.tight_layout()
plt.show()
```
